```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("feature_scaling_dataset (1).csv")
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
sns.kdeplot(df['Age'], color='blue')
sns.kdeplot(df['Salary'], color='green')
sns.kdeplot(df['Spending Score'], color='red')
sns.kdeplot(df['Savings'], color='black')

plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
df_scaled=pd.DataFrame(df_scaled, columns=df.columns)
df_scaled
```

```python
np.round(df_scaled.describe(), 1)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 5))

sns.kdeplot(df['Age'], color='blue', ax=axs[0], label="Age")
sns.kdeplot(df['Salary'], color='green',ax=axs[0],label="Salary")
sns.kdeplot(df['Spending Score'], color='red',ax=axs[0],label='Spending Score')
sns.kdeplot(df['Savings'], color='black',ax=axs[0],label='Savings')
axs[0].set_title("Before Scaling")
axs[0].legend(title='Features')

sns.kdeplot(df_scaled['Age'], color='blue', ax=axs[1],label="Age")
sns.kdeplot(df_scaled['Salary'], color='green', ax=axs[1],label="Salary")
sns.kdeplot(df_scaled['Spending Score'], color='red', ax=axs[1],label='Spending Score')
sns.kdeplot(df_scaled['Savings'], color='black', ax=axs[1],label='Savings')
axs[1].set_title("After Scaling")
axs[1].legend(title='Features')

plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
normscaler = MinMaxScaler()
df_normscaled = normscaler.fit_transform(df)
df_normscaled=pd.DataFrame(df_normscaled, columns=df.columns)
df_normscaled
```

```python
np.round(df_normscaled.describe(),1)
```

```python
fig, axs = plt.subplots(1,2,figsize=(15,5))

axs[0].set_title("Before Normalisation")
sns.kdeplot(df['Age'],ax=axs[0],color='blue',label='Age')
sns.kdeplot(df['Salary'],ax=axs[0],color='red',label='Salary')
sns.kdeplot(df['Spending Score'],ax=axs[0],color='green',label='Spending')
sns.kdeplot(df['Savings'],ax=axs[0],color='black',label='Savings')

axs[1].set_title("After Normalisation")
sns.kdeplot(df_normscaled['Age'],ax=axs[1],color='blue',label='Age')
sns.kdeplot(df_normscaled['Salary'],ax=axs[1],color='red',label='Salary')
sns.kdeplot(df_normscaled['Spending Score'],ax=axs[1],color='green',label='Spending')
sns.kdeplot(df_normscaled['Savings'],ax=axs[1],color='black',label='Savings')

plt.tight_layout()
plt.show()
```

# **Standardisation Vs Normalisation**

```python
fig, axs = plt.subplots(1,2,figsize=(15,5))

sns.kdeplot(df_scaled['Age'], color='blue', ax=axs[0],label="Age")
sns.kdeplot(df_scaled['Salary'], color='green', ax=axs[0],label="Salary")
sns.kdeplot(df_scaled['Spending Score'], color='red', ax=axs[0],label='Spending Score')
sns.kdeplot(df_scaled['Savings'], color='black', ax=axs[0],label='Savings')
axs[0].set_title("Standardisation")
axs[0].legend(title='Features')

axs[1].set_title("Normalisation")
sns.kdeplot(df_normscaled['Age'],ax=axs[1],color='blue',label='Age')
sns.kdeplot(df_normscaled['Salary'],ax=axs[1],color='green',label='Salary')
sns.kdeplot(df_normscaled['Spending Score'],ax=axs[1],color='red',label='Spending')
sns.kdeplot(df_normscaled['Savings'],ax=axs[1],color='black',label='Savings')
axs[1].legend(title='Features')

plt.tight_layout()
plt.show()
```

**Observations**
Standardisation scales the value down and centres around the mean of the data.
Normalisation on the other hand, scales the value down in the range of [0,1]. 
All the feature datapoints which were present are now comparable. 
Standardisation is generally useful when the dataset is highly non-uniform, while normalisation is useful when datapoints are associated with units or when min and max range is already provided.
